```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)

path_2022 = '../data/2022/'
path = '../data/'
```

```python
df_ost = pd.read_csv(path+'2_ost.csv', sep=';')
```

```python
df_ost[df_ost['id_tov'] == 49220]
```

```python
id_tov = list(df_ost['id_tov'].unique())
```

```python
tov = {}
for el in id_tov:
    df = df_ost[df_ost['id_tov'] == el]
    tov[el] = {
        'name' : df['name_tov'].iloc[0],
        'name_gr2' : df['name_gr2'].iloc[0],
        'name_gr' : df['name_gr'].iloc[0]
    }
```

```python
pd.DataFrame(tov).T.to_csv(path+'inf_tovar.csv', sep=';')
```
