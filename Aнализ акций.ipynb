```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)

path_2022 = '../data/2022/'
path = '../data/'

FONT_TITLE = 25
FONT_LABEL = 20
```

```python
df_05_svao = pd.read_csv(path_2022+'svao_5_may.csv', sep=';')
df_06_svao = pd.read_csv(path_2022+'svao_6_june.csv', sep=';')
```

```python
df_05_svao['BaseSum_correct'] = df_05_svao['BaseSum'] * df_05_svao['znak']
df_05_svao['Quantity_correct'] = df_05_svao['Quantity'] * df_05_svao['znak']
df_05_svao['date_ch'] = pd.to_datetime(df_05_svao['date_ch'])
df_05_svao = df_05_svao.set_index(pd.DatetimeIndex(df_05_svao['date_ch']))
```

```python
df_06_svao['BaseSum_correct'] = df_06_svao['BaseSum'] * df_06_svao['znak']
df_06_svao['Quantity_correct'] = df_06_svao['Quantity'] * df_06_svao['znak']
df_06_svao['date_ch'] = pd.to_datetime(df_06_svao['date_ch'])
df_06_svao = df_06_svao.set_index(pd.DatetimeIndex(df_06_svao['date_ch']))
```

22817 - Сыр "Чеддар" швейцарский нарезка

```python
a22817_12_18 = df_06_svao[df_06_svao['id_tov_cl'] == 22817]
```

```python
a22817_12_18['id_tt_cl'].value_counts()[:10]
```

```python
plt.figure(figsize=(20,10))
plt.title('Акция по товару 22817 - Сыр "Чеддар" швейцарский нарезка', fontsize=FONT_TITLE)
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==14330]['Quantity'].resample('d').sum(), label='14330')
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==12618]['Quantity'].resample('d').sum(), label='12618')
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==11178]['Quantity'].resample('d').sum(), label='11178')
plt.vlines(x=a22817_12_18['date_ch'].loc['2022-06-12'], ymin=0, ymax=6, linewidth=4.0)
plt.vlines(x=a22817_12_18['date_ch'].loc['2022-06-18'], ymin=0, ymax=6, linewidth=4.0)
plt.ylabel('Количество проданного товара', fontsize=FONT_LABEL)
plt.xticks(rotation=90)
plt.legend( fontsize=FONT_LABEL)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title('Акция по товару 22817 - Сыр "Чеддар" швейцарский нарезка', fontsize=FONT_TITLE)
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==14902]['Quantity'].resample('d').sum(), label='14902')
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==11398]['Quantity'].resample('d').sum(), label='11398')
plt.plot(a22817_12_18[a22817_12_18['id_tt_cl']==12612]['Quantity'].resample('d').sum(), label='12612')
plt.vlines(x=a22817_12_18['date_ch'].loc['2022-06-12'], ymin=0, ymax=6, linewidth=4.0)
plt.vlines(x=a22817_12_18['date_ch'].loc['2022-06-18'], ymin=0, ymax=6, linewidth=4.0)
plt.ylabel('Количество проданного товара', fontsize=FONT_LABEL)
plt.xticks(rotation=90)
plt.legend( fontsize=FONT_LABEL)
plt.show()
```
