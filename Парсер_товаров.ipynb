```python
def add_tovar(soup, tov, tovar_collection):
    try: name_tov = soup.find('nav', class_="Breadcrumbs js-main-breadcrumbs active").find_all('span', class_="Breadcrumbs__link")[-1].text.replace('/', '').strip()
    except: name_tov = None
        
    try: name_gr2 = soup.find('nav', class_="Breadcrumbs js-main-breadcrumbs active").find_all('span', class_="Breadcrumbs__link")[-2].text.replace('/', '').strip()
    except: name_gr2 = None
    
    try: name_gr = soup.find('nav', class_="Breadcrumbs js-main-breadcrumbs active").find_all('span', class_="Breadcrumbs__link")[-3].text.replace('/', '').strip()
    except: name_gr = None
     
    tovar_collection.append({
        'id_tov': tov,
        'name_tov': name_tov,
        'name_gr2': name_gr2,
        'name_gr': name_gr
    })   
    
    
```
