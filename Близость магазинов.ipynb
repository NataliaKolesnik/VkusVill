```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)

path_2022 = '../data/2022/'
path = '../data/'
```

```python
svod_shop = pd.read_csv(path_2022+'svod_shop.csv', sep=';')
```

```python
svod_shop = svod_shop.rename(columns={'Unnamed: 0':'id_tt_cl'})
svod_shop = svod_shop.set_index('id_tt_cl')
svod_shop
```

```python
svod_shop_vv = svod_shop[svod_shop['format']=='ВкусВилл']
```

```python
svod_shop_vv = pd.get_dummies(svod_shop_vv, columns=['Район'])
```

```python
svod_shop_vv
```

```python
svod_shop_vv['shop_list'] = svod_shop_vv['shop_list'].apply(lambda s: s.split(',') if type(s)==str else [])
```

```python
shop = {}

for list_out in svod_shop_vv['shop_list'].to_list():
    for list_inner in list_out:
        shop.setdefault(list_inner, 0)
        shop[list_inner] +=1 
```

```python
dict(pd.Series(shop).sort_values(ascending=False)[:25])
```

```python
svod_shop_vv['Пятёрочка'] = 0
svod_shop_vv['Перекрёсток'] = 0
svod_shop_vv['Магнит'] = 0
svod_shop_vv['Магнолия'] = 0
svod_shop_vv['Дикси'] = 0
svod_shop_vv['Красное & белое'] = 0
svod_shop_vv['Верный'] = 0
svod_shop_vv['Мини Лента'] = 0
svod_shop_vv['Супер Лента'] = 0
svod_shop_vv['ВкусВилл'] = 0
svod_shop_vv['Продуктовые'] = 0
```

```python
def shop_list_dummies(row):
    list_shop = ['Пятёрочка', 'Перекрёсток', 'Магнит', 'Магнолия', 'Дикси', 'Красное & белое',
                 'Верный', 'Мини Лента','Супер Лента', 'ВкусВилл', ]
    shops = row['shop_list']
    for el in shops:
        if el in list_shop:
            row[el] +=1
        else:
            row['Продуктовые'] += 1
    return row
            
    
```

```python
svod_shop_vv = svod_shop_vv.apply(lambda row: shop_list_dummies(row), axis=1)
```

```python
svod_shop_vv.to_csv(path_2022+'ВкусВилл_свод_полный.csv', sep=';')
```
