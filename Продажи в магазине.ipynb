```python
pd.set_option('display.float_format', '{:.2f}'.format)
#%config InlineBackend.figure_format = 'svg' 

path_2022 = '../data/2022/'
path = '../data/'

FONT_TITLE = 25
FONT_LABEL = 15
```

```python
name_tov = pd.read_csv('../data/inf_tovar.csv', sep=';')
```

```python
svod_shop = pd.read_csv(path_2022+'svod_shop.csv', sep=';')
svod_shop = svod_shop.rename(columns={'Unnamed: 0':'id_tt_cl'})

df_05_svao = pd.read_csv(path_2022+'svao_5_may.csv', sep=';')
df_06_svao = pd.read_csv(path_2022+'svao_6_june.csv', sep=';')
```

```python
# Корректируем возвраты, чтоб они отнимались
df_05_svao = adjust_sales(df_05_svao)
df_06_svao = adjust_sales(df_06_svao)
```

```python
for tt in [11127, 14258, 11178, 11357, 15714, 11956, 11271, 11591, 13968, 12046]:
    df_5 = df_05_svao[df_05_svao['id_tt_cl']==tt]
    df_6 = df_06_svao[df_06_svao['id_tt_cl']==tt]
    
    df_5 = df_5.groupby('id_tov_cl', as_index=False).agg({'Quantity':'sum', 'BaseSum':'sum'}).rename(columns={'Quantity':'кол_май', 'BaseSum':'сумма_май'})
    df_6 = df_6.groupby('id_tov_cl', as_index=False).agg({'Quantity':'sum', 'BaseSum':'sum'}).rename(columns={'Quantity':'кол_июнь', 'BaseSum':'сумма_июнь'})
    
    df = df_6.merge(df_5, how='outer', on='id_tov_cl')
    df = df.merge(name_tov, how='left', on='id_tov_cl')
    
    df = df.sort_values(['name_gr', 'name_gr2', 'name'])
    
#     df.to_csv(path_2022+f'продажи/{tt}_Продажи_товары.csv', sep=';', index=False)
    
    # сгруппируем по группам
#     df_name_gr = df.groupby('name_gr', as_index=False).agg({'кол_июнь':['count','sum'], 'сумма_июнь':'sum', 'кол_май':['count','sum'], 'сумма_май':'sum'})\
    
#     ploshad = svod_shop[svod_shop['id_tt_cl']==tt]['ploshad'].values[0]
    
#     df_name_gr['площадь'] = ploshad
#     df_name_gr.to_csv(path_2022+f'продажи/{tt}_Продажи_группы.csv', sep=';', index=False)
    
    # сгруппируем по подгруппам
    df_name_gr2 = df.groupby('name_gr2', as_index=False).agg({'кол_июнь':['count','sum'], 'сумма_июнь':'sum', 'кол_май':['count','sum'], 'сумма_май':'sum'})\
    
    ploshad = svod_shop[svod_shop['id_tt_cl']==tt]['ploshad'].values[0]
    
    df_name_gr2['площадь'] = ploshad
    df_name_gr2.to_csv(path_2022+f'продажи/{tt}_Продажи_подгруппы.csv', sep=';', index=False)
```
