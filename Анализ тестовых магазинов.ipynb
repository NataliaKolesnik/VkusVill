```python
list_fname_2022 = ['svao_2_febr.csv', 'svao_3_march.csv', 'svao_4_april.csv', 'svao_5_may.csv', 'svao_6_june.csv'] 
list_month = ['февр', 'март', 'апр', 'май', 'июнь']
```

```python
for id_tt in test_shop_sh:
    
    name_file_6 = f'{id_tt}_6_июнь.csv'
    df_tof = pd.read_csv(path_2022+name_file_6, sep=';')
    tov_list = df_tof.id_tov_cl.unique()
    
    df_prod_2022 = pd.DataFrame(tov_list, columns=['id_tov_cl'])

    for i in range(len(list_fname_2022)):
        
        df = pd.read_csv(path+list_fname_2022[i], sep=';')
        
        df['Quantity'] = df['Quantity'] * df['znak']
        df['BaseSum'] = df['BaseSum'] * df['znak']
        
        df = df[df['id_tt_cl'] == id_tt].groupby(['id_tov_cl'], as_index=False)\
             .agg({'Quantity': 'sum', 'BaseSum':'sum'})\
             .rename(columns={'Quantity':f'Кол_{list_month[i]}', 'BaseSum':f'Сумма_{list_month[i]}'})
        
        df_prod_2022 = df_prod_2022.merge(df, how='outer', on='id_tov_cl')


    df_prod_2022 = df_prod_2022.fillna(0)

    df_prod_2022['Кол_22'] = (df_prod_2022['Кол_февр'] + df_prod_2022['Кол_март'] + df_prod_2022['Кол_апр'] 
                              + df_prod_2022['Кол_май'] + df_prod_2022['Кол_июнь'])

    df_prod_2022['Сумма_22'] = (df_prod_2022['Сумма_февр'] + df_prod_2022['Сумма_март'] + df_prod_2022['Сумма_апр']
                                + df_prod_2022['Сумма_май'] + df_prod_2022['Сумма_июнь'])


    df_prod_2022 = df_prod_2022.merge(inf_tovar, how='left', on='id_tov_cl')\
                               .merge(df_tof[['id_tov_cl', 'Quantity_median', 'Quantity_mean', 'BaseSum_median', 
                                              'BaseSum_mean', 'Shop_count', 'Shop_list']], how='left', on='id_tov_cl')

    df_prod_2022 = df_prod_2022.fillna(0)



    for col in ['Кол_февр', 'Сумма_февр', 'Кол_март', 'Сумма_март', 'Кол_апр', 'Сумма_апр', 
                'Кол_май', 'Сумма_май', 'Кол_июнь', 'Сумма_июнь', 'Кол_22', 'Сумма_22']:
        df_prod_2022[col] = df_prod_2022[col].apply(lambda s: np.round(s, 2))
        
    df_prod_2022.to_csv(path+f'{id_tt}_22.csv', sep=';', index=False)    
```
